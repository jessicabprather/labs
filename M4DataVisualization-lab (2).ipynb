## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')

QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)

df['ConvertedComp'].plot(kind='hist', figsize=(8, 5))
plt.title('Histogram of Converted Compensation')
plt.ylabel('Number of Respondents')
plt.xlabel('Annual USD')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
df['Age'].plot(kind='box', figsize=(4,5))
plt.title('Age of Respondents Box Plot')
plt.show()
print(df['Age'].describe())
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import seaborn as sns
plot=sns.scatterplot(x="Age", data=df, y="WorkWeekHrs")
plt.title('Correlation Between Age and Hours Worked')
print('correlation: ',df["Age"].corr(df["WorkWeekHrs"]))
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
#sns.scatterplot(data=df,x="WorkWeekHrs",y="CodeRevHrs",size="Age")
sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5, sizes=(10, 500)) 
plt.title('Code Review Hours per Weekly Hours by Age')
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT * FROM DatabaseDesireNextYear
"""
df_db = pd.read_sql_query(QUERY,conn)

# group respondents by database and count
df_groupby = df_db.groupby('DatabaseDesireNextYear', axis=0).count()

df_groupby.sort_values(["Respondent"], ascending=False, inplace=True)

df_pie = df_groupby.head()
 
df_pie['Respondent'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 Databases To Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY,conn)

#group by age and take median values
df_chart = df.groupby('Age').median()

df_chart.plot(kind='bar', stacked=True)

plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs Ages 30 to 35')
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age
FROM master where Age between 45 and 60
"""
df = pd.read_sql_query(QUERY,conn)

df_line = df.groupby('Age').median()

df_line.plot(kind='line')

plt.title('Median ConvertedComp for Ages 45 to 60')
#plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, COUNT(*) as MainBranch from master
group by MainBranch
"""

df=pd.read_sql_query(QUERY,conn)

df.plot(kind='barh')
plt.title('Main Branch')
plt.show()
```
